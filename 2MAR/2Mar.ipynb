Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)

plt.title('Relation between X and Y')
plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

plt.subplots():

The subplot() function in Matplotlib is used to create multiple plots in a single figure. By sepecifing the no. of rows and columns.

```python
# For line 1:
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# For line 2:
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# For line 3:
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# For line 4:
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating a figure with 2 rows and 2 columns
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))

# Creating the first subplot
ax1.plot(x1, y1)
ax1.set_title('Line 1')

# Creating the second subplot
ax2.plot(x2, y2)
ax2.set_title('Line 2')

# Creating the third subplot
ax3.plot(x3, y3)
ax3.set_title('Line 3')

# Creating the fourth subplot
ax4.plot(x4, y4)
ax4.set_title('Line 4')

# Displaying the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

plt.bar():

It is used to plot numerical values of different categories. Bar plots are useful in comparing different categories or groups of data and identifying trends or patterns in the data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.show()
```

```python
plt.barh(company, profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

plt.boxplot()

Box plot is a graphical representation of the distribution of data based on five number summary: minimum, first quartile, median, third quartile, and maximum.
It is commonly used for identifying outliers and helps to understanding the spread and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1, box2])
plt.show()
```
