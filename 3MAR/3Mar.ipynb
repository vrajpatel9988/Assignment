Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(fmri, x='timepoint', y='signal', errorbar=None)
```

```python
sns.lineplot(fmri, x='timepoint', y='signal', hue='region', errorbar=None)
```

```python
sns.lineplot(fmri, x='timepoint', y='signal', hue='event', errorbar=None)
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.boxplot(titanic, x='pclass', y='age')
```

```python
sns.boxplot(titanic, x='pclass', y='fare')
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.histplot(diamonds, x='price', hue='cut')
```

```python
sns.histplot(diamonds, x='price', hue='cut', element='poly')
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.heatmap(flights.drop(columns=['month']))
```

```python
flights_df = flights.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(flights_df, annot=True, fmt='d')
```
